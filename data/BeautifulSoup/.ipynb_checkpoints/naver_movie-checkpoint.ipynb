```python
date = pd.date_range('2019.10.01', periods= 31, freq = 'D')
date
```

```python
import time 

movie_date = []
movie_name = []
movie_point = []
```

```python
for today in date:
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}'
    response = urlopen(html.format(date = today.strftime('%Y%m%d')))
    soup = BeautifulSoup(response, 'html.parser')
    
    movie_date.extend([today]*len(soup.find_all('td','point')))
    movie_name.extend([each.a.string for each in soup.find_all('div', 'tit5')])
    movie_point.extend([each.string for each in soup.find_all('td', 'point')])
    
    print(str(today))
    time.sleep(0.1)
```

```python
len(movie_date)
```

```python
len(movie_name)
```

```python
len(movie_point)
```

```python
movie = pd.DataFrame({'date':movie_date, 'name': movie_name, 'point' : movie_point})
movie.head()
```

```python
movie.info()
```

```python
movie['point'] = movie['point'].astype(float)
```

```python
movie.to_csv('./naver_movie.csv', sep=',', encoding = 'utf-8')
```

```python
import numpy as np
import pandas as pd 

movie = pd.read_csv("./naver_movie.csv", index_col = 0)
movie.head()
```

```python
movie_unique = pd.pivot_table(movie, index = ['name'], aggfunc=np.sum)
movie_unique.head()
```

```python
movie_best = movie_unique.sort_values(by='point', ascending = False)
movie_best.head(10)
```

```python
movie.query('name==["조커"]')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure(figsize = (12,8))
plt.plot(movie.query('name == ["조커"]')['date'],
         movie.query('name == ["조커"]')['point'])

plt.xticks("")
plt.ylim((5,10))
plt.grid()
plt.show()
```

```python
movie_pivot = pd.pivot_table(movie, index = ['date'], columns = ['name'])
movie_pivot.columns = movie_pivot.columns.droplevel([0])
movie_pivot.head()
```

```python
len(movie_pivot.columns)

```

```python
target_col = ['교회오빠', '82년생 김지영' , '조커' , '기생충' , '터미네이터: 다크 페이트']
plt.figure(figsize = (12,8))
plt.plot(movie_pivot[target_col])
plt.legend (target_col, loc = 'best')
plt.tick_params(bottom = False, labelbottom = False)
plt.show()
```
