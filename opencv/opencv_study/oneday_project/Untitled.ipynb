## 영상 캡쳐 코드 

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='NanumGothic') ## NANUMGOTHIC 전역 폰트로 설정하는 코드 
plt.rcParams['axes.unicode_minus'] = False
```

```python

cap = cv2.VideoCapture("/dev/video0")
if not cap.isOpened():
    raise RuntimeError("ERROR! Unable to open camera")
try:
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    heigth = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    print(f'width = {width}, heigth = {heigth}')
    fourcc = cv2.VideoWriter_fourcc(*'DIVX')
    out = cv2.VideoWriter('./test__.avi', fourcc, 20,(width, heigth), isColor=False)
    cnt = 1
    
    while True:
        k = cv2.waitKey(1) & 0xFF
        ret , frame = cap.read()
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame', gray)
        out.write(gray)
            
        if k == ord('s'):
            print("screenshot saved")
            cv2.imwrite('./screenshot{}.png'.format(cnt), frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
            cnt += 1
                
        elif k == 27 : break
                
            
finally:
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```

# 녹화중 화면에 뛰우기 

## 녹화중 이미지 만들기

```python
import cv2
cap = cv2.VideoCapture("/dev/video0")
if not cap.isOpened():
    raise RuntimeError("ERROR! Unable to open camera")
try:
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    heigth = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    print(f'width = {width}, heigth = {heigth}')
    fourcc = cv2.VideoWriter_fourcc(*'DIVX')
    out = cv2.VideoWriter('./test__.avi', fourcc, 20,(width, heigth), isColor=False)
    cnt = 1
    flag = False
    
    
    
    while True:
        k = cv2.waitKey(1) & 0xFF
        ret , frame = cap.read()
        
        new = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        if flag:
            cv2.imshow('frame', new)
        else:
            cv2.imshow('frame', gray)
            
########################################
        
        if flag:
            out.write(gray)
            
#######################################
            
        if k == ord('s'):
            print("screenshot saved")
            cv2.imwrite('./screenshot{}.png'.format(cnt), frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
            cnt += 1
            
        elif k == ord('v'):
            print('녹화중')
            flag = not flag
            
            
        elif k == 27 : break
                
            
finally:
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```

# 단추키 설명 오버레이 

```python
import cv2
cap = cv2.VideoCapture("/dev/video0")
if not cap.isOpened():
    raise RuntimeError("ERROR! Unable to open camera")
try:
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    heigth = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    print(f'width = {width}, heigth = {heigth}')
    fourcc = cv2.VideoWriter_fourcc(*'DIVX')
    out = cv2.VideoWriter('./test__.avi', fourcc, 20,(width, heigth), isColor=False)
    cnt = 1
    flag = False
    trigger = False
    
    img = np.zeros((480,640,3), np.uint8)
    font = cv2.FONT_HERSHEY_COMPLEX
    img = cv2.putText(img, text = 's is screenshot', org= (100,100), fontFace=font, fontScale = 1, color=(255,0,0),thickness = 4)
    img = cv2.putText(img, text = 'v is recording', org= (100,200), fontFace=font, fontScale = 1, color=(255,0,0),thickness = 4)
    img = cv2.putText(img, text = 'h is mannual', org= (100,300), fontFace=font, fontScale = 1, color=(255,0,0),thickness = 4)
    
    
    while True:
        k = cv2.waitKey(1) & 0xFF
        ret , frame = cap.read()
        
        new = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        
        
############################################        
        if trigger:
            new_blended = cv2.addWeighted(src1= new, alpha = 0.5, src2 = img, beta = 0.5, gamma =0)
            gray_blended = cv2.addWeighted(src1= gray, alpha = 0.5, src2 = img, beta = 0.5, gamma =0)
############################################            
        
        if flag:
            cv2.imshow('frame', new)
            if trigger:
                cv2.imshow('frame', new_blended)
        else:
            cv2.imshow('frame', gray)
            if trigger:
                cv2.imshow('frame',gray_blended)
            
########################################
        
        if flag:
            out.write(gray)
            
#######################################
            
        if k == ord('s'):
            print("screenshot saved")
            cv2.imwrite('./screenshot{}.png'.format(cnt), frame, params=[cv2.IMWRITE_PNG_COMPRESSION,0])
            cnt += 1
            
        elif k == ord('v'):
            print('녹화중')
            flag = not flag
        
        elif k == ord('h'):
            print('오버레이')
            trigger = not trigger 
            
            
        elif k == 27 : break
                
            
finally:
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```
