# 회전

```python
import math 
import sys
import cv2
import numpy as np

src = cv2.imread('../data/gorilla.jpg')

if src is None:
    print('image load failed')
    sys.exit()
    
    
rad = 20 * math.pi / 180 ## rad -> degree

aff = np.array([[math.cos(rad), math.sin(rad), 0], [-math.sin(rad), math.cos(rad), 0]], dtype=np.float32)

dst = cv2.warpAffine(src, aff,(0,0))

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()

## 회전은 원점을 기준으로 함을 명심하자 
```

### 원점이 아닌 곳을 기준으로 회전할 때 

```python
import math 
import sys
import cv2
import numpy as np

src = cv2.imread('../data/gorilla.jpg')

if src is None:
    print('image load failed')
    sys.exit()

cp = (src.shape[1]/2 , src.shape[0]/2) ## y,x 의 중점 

rot = cv2.getRotationMatrix2D(cp, 20, 0.5)
## 회전행렬 구하기 

dst = cv2.warpAffine(src, rot, (0,0))

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

## 어파인 변환과 투시 변환

```python
import sys
import numpy as np
import cv2

src = cv2.imread('../data/pinkwink_namecard.jpeg')

if src is None:
    print('image load failed')
    sys.exit()
    
w, h = 720, 400
srcQuad = np.array([[360,345], [879,404], [895,664], [254,573]], np.float32)
dstQuad= np.array([[0,0], [w-1,0], [w-1,h-1], [0,h-1]], np.float32)

pers = cv2.getPerspectiveTransform(srcQuad, dstQuad)
## 변환행렬 구하는 함수 
dst = cv2.warpPerspective(src, pers, (w,h))
## 변환하기 
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()

## 전체 이미지에서 원하는 그림 , 원하는 부분의 픽셀만 가져올때 유옹하게 사용할 수 있음
```

### 왜곡된 문서 똑바로 펴기 (투시변환)

```python
import sys
import numpy as np 
import cv2

def drawROI(img , corners):
    cpy = img.copy() ## 이미지 객체 만들기 
    
    c1 = (192, 192, 255)
    c2 = (128, 128, 255) ## 색상 지정 
    
    for pt in corners:
        cv2.circle(cpy, tuple(pt.astype(int)), 25, c1, -1 , cv2.LINE_AA) ## 매개변수로 입력받은 코너 좌표에 원 그리기 
    
    cv2.line(cpy, tuple(corners[0].astype(int)), tuple(corners[1].astype(int)), c2, 2, cv2.LINE_AA) ## 원 중심끼리 이어주기
    
    cv2.line(cpy, tuple(corners[1].astype(int)), tuple(corners[2].astype(int)), c2, 2, cv2.LINE_AA)
    
    cv2.line(cpy, tuple(corners[2].astype(int)), tuple(corners[3].astype(int)), c2, 2, cv2.LINE_AA)
    
    cv2.line(cpy, tuple(corners[3].astype(int)), tuple(corners[4].astype(int)), c2, 2, cv2.LINE_AA) 
    
    disp = cv2.addWeighted(img, 0.3, cpy, 0.7, 0) ## 원래 이미지랑 원 그린 이미지랑 합치기 
    
    return disp

def onMouse(event, x, y, flags , param):
    global srcQuad, dragSrc, ptOld, src
    
    ## 왼쪽 버튼이 내려갔을때 
    ## 원 중심으로 부터 반경으로 25까지 검출하기  
    if event == cv2.EVENT_LBUTTONDOWN:
        for i in range(4):
            if cv2.norm(srcQuad[i] - (x,y)) < 25:
                dragSrc[i] = True
                ptOld = (x,y)
                break
            
    if event == cv2.EVENT_LBUTTONUP:
        for i in range(4):
            dragSrc[i] = False 
            
    if event == cv2.EVENT_MOUSEMOVE:
        for i in range(4):
            if dragSrc[i]:
                
                ## 마우스 x이동 y이동 변화량 
                dx = x - ptOld[0]
                dy = y - ptOld[1]
                
                #마우스 이동한 만큼 (중심원기준 어디를 잡든) 중심좌표가 정확하게 이동한다. 
                srcQuad[i] += (dx, dy)
                
                cpy = drawROI(src, srcQuad)
                cv2.imshow('img', cpy)
                ptOld = (x,y)
                break






src = cv2.imread('../data/123123.png')

if src is None:
    print('image load failed')
    sys.exit()
    
## 입력 영상 크기 및 출력 영상 크기 
h,w = src.shape[:2]
dw = 500
dh = round(dw * 297 / 210)  ## a4용지크기 : 210 x 297 cm

## 모서리 점들의 좌표, 드래그 상태 flag
## 초기 사각형 사이즈 srcQuad
## 출력 사이즈 dstQuad
srcQuad = np.array([[30,30], [30, h-30], [w-30,h-30], [w-30,30]], np.float32)
dstQuad = np.array([[0,0], [0, dh-1], [dw-1, dh-1], [dw-1,0]], np.float32)
dragSrc = [False, False, False, False]

# 모서리점, 사각형 그리기 
disp = drawROI(src, srcQuad)

cv2.imshow('img', disp)
cv2.setMouseCallback('img' , onMouse)

while True:
    key = cv2.waitKey()
    if key == 13: ## enter키
        break
    ## 엔터키를 누르면 무한  루프를 빠져나가 밑에 코드르 실행하도록 코드를 구성
    
    elif key == 27: ## esc키 
        cv2.destroyAllWindows('img')
        sys.exit()
    
## 투시변환
pers = cv2.getPerspectiveTransform(srcQuad, dstQuad)
dst = cv2.warpPerspective(src, pers, (dw,dh), flags = cv2.INTER_CUBIC)

## 결과 영상 출력
cv2.imshow('dst', dst)

while True:
    if cv2.waitKey() == 27:
        break
    
cv2.destroyAllWindows()
```
