```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/rainbow.jpg')
print('ftype(img) = {type(img)}, img.shape={img.shape}')
plt.imshow(img)
```

```python
img = cv2.imread('../data/rainbow.jpg' , cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
print(ret)

## THRESH_BINARY ret  255  0 binary image 
```

```python
plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY_INV)
print(ret)

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()

## INV BINARY  REVERSE  
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_TRUNC)
print(ret)

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()

## THRESH_TRUNC ret   255         
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_TOZERO_INV)
print(ret)

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()
```

```python
img = cv2.imread('../data/crossword.jpg', 0)
print(f'type(img) = {type(img)}, img.shape = {img.shape}')
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    
## 
```

```python
show_pic(img)
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
show_pic(thresh1)
```

```python
ret , thresh1 = cv2.threshold(img, 180,255,cv2.THRESH_BINARY)
show_pic(thresh1)
```

```python
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11 ,8)
show_pic(thresh2)

## adaptiveThreshold     adaptive_thresh    
##    binary   
```

```python
blended = cv2.addWeighted(src1 = thresh1, alpha = 0.6, src2 = thresh2, beta = 0.4, gamma = 0)
show_pic(blended)

##       addWeighted 
```

```python
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11 ,8)
show_pic(thresh2)
```
