```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/rainbow.jpg')
print('ftype(img) = {type(img)}, img.shape={img.shape}')
plt.imshow(img)
```

```python
img = cv2.imread('../data/rainbow.jpg' , cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
print(ret)

## THRESH_BINARY는 ret값보다 큰것은 255 작은것은 0으로 binary image를 생성
```

```python
plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY_INV)
print(ret)

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()

## INV는 BINARY이미지 값을 REVERSE하는 것 
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_TRUNC)
print(ret)

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()

## THRESH_TRUNC함수는 ret값보다 큰 값은 255로 만드나 작은 값은 본인 픽셀 값 그대로 유지 
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_TOZERO_INV)
print(ret)

plt.subplot(1,2,1)
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(thresh1, cmap = 'gray')
plt.show()
```

```python
img = cv2.imread('../data/crossword.jpg', 0)
print(f'type(img) = {type(img)}, img.shape = {img.shape}')
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    
## 
```

```python
show_pic(img)
```

```python
ret , thresh1 = cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
show_pic(thresh1)
```

```python
ret , thresh1 = cv2.threshold(img, 180,255,cv2.THRESH_BINARY)
show_pic(thresh1)
```

```python
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11 ,8)
show_pic(thresh2)

## adaptiveThreshold는 작은 박스 사이즈에 해당하는만큼 adaptive_thresh 픽셀값의 평균을 냄 
## 그리고 그 출력을 binary 결과로 보냄 
```

```python
blended = cv2.addWeighted(src1 = thresh1, alpha = 0.6, src2 = thresh2, beta = 0.4, gamma = 0)
show_pic(blended)

## 두 이미지 가중치 두어 합치는 함수 addWeighted함수 
```

```python
thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11 ,8)
show_pic(thresh2)
```
