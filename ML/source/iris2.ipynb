```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

features = iris.data[:, 2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.2, random_state= 13)
```

```python
import numpy as np
```

```python
np.unique(y_test, return_counts=True)
## 동일 비율이 아니다. 
```

```python
features = iris.data[:, 2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.2, stratify = labels, random_state= 13)
```

```python
np.unique(y_test, return_counts=True)
## stratify 옵션을 사용하면 테스트 모델을 동일 비율로 나눌 수 있다.
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)

## 학습의 일관성을 위한 random_state
## 모델 단순화를 위한 max_depth
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(12,8))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree,  legend =2)
plt.show()
```

```python
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

# Zip 과 언패킹

```python
list1 =  ['a', 'b', 'c']
list2 = [1,2,3]
```

```python
A = zip(list1, list2)
dict(A)
## zip 한 이후에 어떻게 항목 자료형을 할지 정해주면 된다. 
```

```python
pairs = [pair for pair in zip(list1, list2)]
pairs

## 리스트 두개를 튜플로 zip
```

```python
pairs = [list(pair) for pair in zip(list1, list2)]
pairs
# ## 리스트 두개를 리스트로 zip 
```

```python
dict(pairs)
```

```python
dict(zip(list1,list2))
```

```python
list(zip(list1,list2))
```

```python
a,b = zip(*pairs)
## 리스트 내부 리스트 요소를 분리할때 zip 함수에 * 인자를 사용하면 된다. 
```
