```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def sin_wave(amp,freq,time):
    return amp*np.sin(2*np.pi*freq*time)
```

```python
time = np.arange(0, 10 , 0.001)
sin1 = sin_wave(1,10,time)
sin2 = sin_wave(2,5,time)
sin3 = sin_wave(4,1,time)
```

```python
plt.figure(figsize = (12,5))
plt.plot(time, sin1, label = r"$\sin {20\pi} t$", color = 'red')
plt.plot(time, sin2, label = r"$\sin {10\pi} t$", color = 'blue')
plt.plot(time, sin3, label = r"$\sin {2\pi} t$", color = 'green')
plt.legend(); plt.grid(); plt.show()
```

### 삼각함수 신호 3개의 합 

```python
sin_sum = sin1 + sin2 + sin3

plt.figure(figsize=(12,5))
plt.plot(time, sin_sum)
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (12,7))
plt.plot(time, sin1, label = r"$\sin {20\pi} t$", color = 'green', alpha = 0.5)
plt.plot(time, sin2, label = r"$\sin {10\pi} t$", color = 'blue' , alpha = 0.5)
plt.plot(time, sin3, label = r"$\sin {2\pi} t$", color = 'black' , alpha = 0.5)
plt.plot(time, sin_sum , label = 'sum_of_sin', color = 'black', lw=0.8)
plt.legend(); plt.grid(); plt.show()
```

```python
n = len(sin_sum)
k = np.arange(n)
Fs = 1/0.001
T = n/Fs
freq = k/T
freq = freq[range(int(n/2))]
```

```python
Y = np.fft.fft(sin_sum)/n
Y = Y[range(int(n/2))]
```

## 정현파의 합성으로 신호를 설명하는  것을 주파수 분석이라  함 
* 주파수의 신호의 합성을 보고 분리 및 분석하는 것 
* 푸리에 변환은 시간영역 신호에서 주파수 성분!을 찾아준다. 

```python
fig, ax = plt.subplots(2,1, figsize=(12,8))
ax[0].plot(time, sin_sum)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude'); ax[0].grid(True)
ax[1].plot(freq,abs(Y), 'r', linestyle='', marker='^')
ax[1].set_xlabel('Freq(Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[1].vlines(freq,[0],abs(Y))
ax[1].set_xlim([0,20]); ax[1].grid(True)
plt.show()
```

## np.concatenate 함수는 단순히 붙여주는 기능을 함 

```python
sin_concat = np.concatenate((sin1, sin2, sin3, sin_sum))
```

```python
time = np.arange(0,40,0.001)

plt.figure(figsize=(12,5))
plt.plot(time, sin_concat)
plt.grid()
plt.show()
```

```python
n = len(sin_concat)
k = np.arange(n)
Fs = 1/0.001; T=n/Fs
freq = k/T
freq = freq[range(int(n/2))]
Y = np.fft.fft(sin_concat)/n
Y = Y[range(int(n/2))]
```

```python
fig, ax = plt.subplots(2,1, figsize=(12,8))
ax[0].plot(time, sin_concat)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude'); ax[0].grid(True)
ax[1].plot(freq,abs(Y), 'r', linestyle='', marker='^')
ax[1].set_xlabel('Freq(Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[1].vlines(freq,[0],abs(Y))
ax[1].set_xlim([0,20]); ax[1].grid(True)
plt.show()
```

## 일정 기간을 잘라서 구간별 ttf를 수행하는 것을 stft라  함
stft 분석을 하는 이유 
실제 주파수는 시간 축을 두고 다양한 주파수가 겹쳐서 하나의 주파수를 형성할 것임 

그런데 이것을 주파수 전체로 분석하여 보면 
주파수 분석은 x 축이 주파수이기 때문에 분석이 불가능함. 의미가 불분명해짐

이를 구분하기 위해 
구간별로 잘라서 주파수를 분석하여 하나의 주파수를 개별로 분석하기 위함 
이를 도와주는 것이 stft 

```python
def draw_stft(f,t,Zxx):
    plt.figure(figsize=(12,5))
    plt.pcolormesh(t,f,np.abs(Zxx), vmin = 0, vmax = 1, shading = 'gouraud')
    plt.title('STFT Magnitude'); plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time[Sec]'); plt.ylim([0,20]); plt.show()
```

```python
from scipy import signal
```

```python
def calc_stft(nperseg):
    f,t,Zxx = signal.stft(sin_concat, Fs, nperseg=nperseg)
    draw_stft(f,t,Zxx)
```

```python
calc_stft(500)
```

```python
calc_stft(1000)
```

```python
calc_stft(2000)
```

```python
calc_stft(5000)
```
