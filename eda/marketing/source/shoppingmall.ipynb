```python
tmp = (df['노출수'].sort_values())/1000
tmp = tmp.reset_index().drop('index', axis = 1)
## 전체 데이터 프레임에서 하나의 칼럼만 추려내어 작은 데이터 프레임 만드는 방법
tmp
```

```python
plt.figure(figsize=(13,5))
plt.plot(tmp['노출수'])
plt.yticks([0,2000,4000,6000,8000,10000], ['0', '2M', '4M', '6M', '8M','10M'])
## Y label 재정의할때 
plt.title("노출수 plot", fontsize = 20)
plt.show()
```

```python
df['클릭수'] = df['클릭수'].astype(int)
## 칼럼 데이터 타입 변환  astype 함수 
```

```python
df.head()
```

```python
df['클릭률(%)'] = df['클릭수'] / df['노출수'] * 100
```

```python
df.head()
```

```python
cpc = round(df['평균클릭비용(VAT포함,원)'], 0)
##cpc 변수 생성
df['평균클릭비용(VAT포함,원)'] = cpc.astype(int)
# 변수를 생성하여 데이터를 다시 만든다음에 칼럼에 대입
df.head()
```

```python
df['광고그룹'].value_counts()
```

```python
imp = df['노출수']
imp = imp[imp >=imp.quantile(0.95)]
# imp 변수를 노출수 칼럼 데이터를 갖는
# imp 상위 5% 이상 노출수를 가진 아이템만 true로 bool 마스킹후 true인 열만 가져옴
imp
```

```python
df_index = df.set_index('키워드')
imp = df_index['노출수']
clk = df_index['클릭수']
result = df_index[(imp >imp.quantile(0.95)) & (clk>=clk.quantile(0.95))]

# df_index라는 키워드 칼럼을 index한 데이터 프레임 재설정 
# imp는 df_index의 노출수 칼럼 데이터 series
# clk는 df_index의 클릭수 칼럼 데이터 series
# 
```
